```python
#read the csv into dataframe
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/bulas/df_bulas_split.csv')
```

```python
#As embeddings column is saved as a text here we convert it to an array so we can later apply the search methods
def split_by_space(string_array):
  new_array = string_array.replace('[','').replace('array','').replace(']','').replace('(','').replace(')','').replace('\n','')
  new_array = new_array.split(', ')
  new_array = [float(i) for i in new_array]
  return new_array

df['embeddings'] = df['embeddings'].apply(lambda x : split_by_space(x))
```
